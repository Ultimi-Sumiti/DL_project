```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
import matplotlib.pyplot as plt
import os
import pretty_midi # You will need to install this library: pip install pretty_midi

# --- Configuration (unchanged) ---
config = {
    'latent_dim': 100,
    'image_height': 128, # Number of notes (e.g., all 128 MIDI notes)
    'image_width': 16,  # Number of timesteps in a bar (e.g., 16th notes per bar)
    'channels': 1,
    'batch_size': 64,
    'num_epochs': 50,
    'lr_g': 0.0002,
    'lr_d': 0.0002,
    'beta1': 0.5,
    'beta2': 0.999,
    'model_save_path': 'outputs/checkpoints/',
    'image_save_path': 'outputs/images/',
    'midi_save_path': 'outputs/midi/', # Added a specific path for MIDI files
    'log_step': 10,
}

# Ensure output directories exist
os.makedirs(config['model_save_path'], exist_ok=True)
os.makedirs(config['image_save_path'], exist_ok=True)
os.makedirs(config['midi_save_path'], exist_ok=True) # Create directory for MIDI files


# --- Device Configuration (unchanged) ---
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# --- Data Loading (Simulated for demonstration - unchanged) ---
def get_dummy_data(num_samples):
    return torch.randint(0, 2, (num_samples, config['channels'], config['image_height'], config['image_width'])).float()

# --- Models (Generator and Discriminator - unchanged from previous correction) ---
class Generator(nn.Module):
    def __init__(self, latent_dim, channels, img_height, img_width):
        super(Generator, self).__init__()
        self.main = nn.Sequential(
            nn.ConvTranspose2d(latent_dim, 512, 4, 1, 0, bias=False),
            nn.BatchNorm2d(512),
            nn.ReLU(True),
            nn.ConvTranspose2d(512, 256, 4, 2, 1, bias=False),
            nn.BatchNorm2d(256),
            nn.ReLU(True),
            nn.ConvTranspose2d(256, 128, 4, 2, 1, bias=False),
            nn.BatchNorm2d(128),
            nn.ReLU(True),
            nn.ConvTranspose2d(128, channels, kernel_size=(8, 1), stride=(8, 1), padding=(0, 0), bias=False),
            nn.Sigmoid()
        )
    def forward(self, input):
        return self.main(input.view(input.size(0), self.main[0].in_channels, 1, 1))

class Discriminator(nn.Module):
    def __init__(self, channels, img_height, img_width):
        super(Discriminator, self).__init__()
        self.main = nn.Sequential(
            nn.Conv2d(channels, 128, kernel_size=(8, 1), stride=(8, 1), padding=(0, 0), bias=False),
            nn.LeakyReLU(0.2, inplace=True),
            nn.Conv2d(128, 256, 4, 2, 1, bias=False),
            nn.BatchNorm2d(256),
            nn.LeakyReLU(0.2, inplace=True),
            nn.Conv2d(256, 512, 4, 2, 1, bias=False),
            nn.BatchNorm2d(512),
            nn.LeakyReLU(0.2, inplace=True),
            nn.Conv2d(512, 1, 4, 1, 0, bias=False),
            nn.Sigmoid()
        )
    def forward(self, input):
        return self.main(input).view(-1, 1).squeeze(1)

# Initialize models (unchanged)
netG = Generator(config['latent_dim'], config['channels'], config['image_height'], config['image_width']).to(device)
netD = Discriminator(config['channels'], config['image_height'], config['image_width']).to(device)

# --- Loss Function and Optimizers (unchanged) ---
criterion = nn.BCELoss()
optimizerD = optim.Adam(netD.parameters(), lr=config['lr_d'], betas=(config['beta1'], config['beta2']))
optimizerG = optim.Adam(netG.parameters(), lr=config['lr_g'], betas=(config['beta1'], config['beta2']))

# Fixed noise for visualization (unchanged)
fixed_noise = torch.randn(config['batch_size'], config['latent_dim'], device=device)

# --- Training Loop (unchanged) ---
G_losses = []
D_losses = []

print("Starting Training Loop...")
for epoch in range(config['num_epochs']):
    real_data = get_dummy_data(config['batch_size']).to(device)
    b_size = real_data.size(0)
    real_labels = torch.full((b_size,), 1.0, dtype=torch.float, device=device)
    fake_labels = torch.full((b_size,), 0.0, dtype=torch.float, device=device)

    netD.zero_grad()
    output_D_real = netD(real_data)
    errD_real = criterion(output_D_real, real_labels)
    errD_real.backward()
    D_x = output_D_real.mean().item()

    noise = torch.randn(b_size, config['latent_dim'], device=device)
    fake_data = netG(noise)
    output_D_fake = netD(fake_data.detach())
    errD_fake = criterion(output_D_fake, fake_labels)
    errD_fake.backward()
    D_G_z1 = output_D_fake.mean().item()

    errD = errD_real + errD_fake
    optimizerD.step()

    netG.zero_grad()
    noise_for_G = torch.randn(b_size, config['latent_dim'], device=device)
    fake_data_for_G = netG(noise_for_G)
    output_G = netD(fake_data_for_G)
    errG = criterion(output_G, real_labels)
    errG.backward()
    D_G_z2 = output_G.mean().item()
    optimizerG.step()

    G_losses.append(errG.item())
    D_losses.append(errD.item())

    if epoch % config['log_step'] == 0 or epoch == config['num_epochs'] - 1:
        print(f"Epoch [{epoch}/{config['num_epochs']}] Loss D: {errD.item():.4f} Loss G: {errG.item():.4f} D(x): {D_x:.4f} D(G(z)): {D_G_z1:.4f} / {D_G_z2:.4f}")

print("Training finished!")

# --- Plotting Training Losses (unchanged) ---
plt.figure(figsize=(10, 5))
plt.title("Generator and Discriminator Loss During Training")
plt.plot(G_losses, label="G")
plt.plot(D_losses, label="D")
plt.xlabel("Iterations (epochs)")
plt.ylabel("Loss")
plt.legend()
plt.grid(True)
plt.show()

# --- Model Saving (unchanged) ---
def save_models(epoch, netG, netD, path_g_dir, path_d_dir):
    os.makedirs(path_g_dir, exist_ok=True)
    os.makedirs(path_d_dir, exist_ok=True)
    
    torch.save(netG.state_dict(), f"{path_g_dir}generator_epoch_{epoch}.pth")
    torch.save(netD.state_dict(), f"{path_d_dir}discriminator_epoch_{epoch}.pth")
    print(f"Models saved at epoch {epoch}")

save_models(config['num_epochs'], netG, netD, config['model_save_path'], config['model_save_path'])

# --- Model Loading Example (unchanged) ---
def load_models(netG_instance, netD_instance, path_g, path_d):
    netG_instance.load_state_dict(torch.load(path_g))
    netD_instance.load_state_dict(torch.load(path_d))
    netG_instance.eval()
    netD_instance.eval()
    print("Models loaded successfully.")
    return netG_instance, netD_instance

# --- Inference Example (Generating new music) ---
print("\nGenerating a new dummy music piece:")
with torch.no_grad():
    z = torch.randn(1, config['latent_dim'], device=device)
    generated_music_piece_tensor = netG(z).cpu()
    print(f"Generated a piece of shape: {generated_music_piece_tensor.shape}")
    print("First 5x5 segment of the generated piano roll (rounded):\n", generated_music_piece_tensor[0, 0, :5, :5].round())

    # --- NEW CODE TO CONVERT TENSOR TO MIDI AND SAVE ---
    # Convert the tensor to a NumPy array for easier processing
    piano_roll_np = generated_music_piece_tensor.squeeze().numpy() # Remove batch and channel dimensions

    # Initialize a PrettyMIDI object
    midi = pretty_midi.PrettyMIDI(initial_tempo=120) # You can adjust tempo

    # Create an instrument (e.g., Acoustic Grand Piano)
    # The 'program' argument determines the instrument sound.
    # You can find a list of program numbers in PrettyMIDI documentation.
    piano = pretty_midi.Instrument(program=pretty_midi.instrument_name_to_program('Acoustic Grand Piano'))

    # Define time resolution. Assuming 16th notes per bar, and 1 bar is 4 beats.
    # So each timestep (width=16) is 0.25 beats if tempo is 120 (1 beat = 0.5s).
    # You might need to adjust this based on your actual data's time resolution.
    time_unit = 0.25 # seconds per timestep (e.g., if a timestep is a 16th note at 120 bpm)
    
    # Iterate through the piano roll to create notes
    # piano_roll_np has shape (image_height, image_width) -> (128, 16)
    for note_idx in range(config['image_height']):
        for time_step in range(config['image_width']):
            # Assuming a binary piano roll where value > 0.5 means note is ON
            if piano_roll_np[note_idx, time_step] > 0.5:
                # pitch: MIDI note number (0-127). note_idx directly maps to pitch.
                # start: start time of the note in seconds
                # end: end time of the note in seconds. For simplicity, assume each note lasts one timestep.
                # velocity: MIDI velocity (0-127). Using a fixed value for simplicity.
                
                pitch = note_idx
                start_time = time_step * time_unit
                
                # To find end time: Look ahead for how long the note is sustained
                duration_timesteps = 1
                for next_time_step in range(time_step + 1, config['image_width']):
                    if piano_roll_np[note_idx, next_time_step] > 0.5:
                        duration_timesteps += 1
                    else:
                        break
                end_time = start_time + (duration_timesteps * time_unit)
                
                # Ensure end_time is greater than start_time
                if end_time > start_time:
                    # You can experiment with different velocities, e.g., 60-100
                    note = pretty_midi.Note(velocity=80, pitch=pitch, start=start_time, end=end_time)
                    piano.notes.append(note)
    
    midi.instruments.append(piano)

    # Save the MIDI file
    output_midi_filename = f"generated_music_epoch_{config['num_epochs']}.mid"
    output_midi_path = os.path.join(config['midi_save_path'], output_midi_filename)
    midi.write(output_midi_path)
    
    print(f"\nGenerated music saved as MIDI to: {output_midi_path}")
    print("You can now open this .mid file with a MIDI player or music software to listen to it.")
```
